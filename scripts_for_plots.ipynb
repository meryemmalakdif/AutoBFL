```python
import matplotlib.pyplot as plt
import numpy as np

# Data
send_rate = np.array([50,91,156,200,240,305,360,405,423])
throughput = np.array([35,70,101.66,106.3,110.73,113.7,114,125,111])
latency = np.array([1.54,1.56,1.78, 2.35 , 2.89,3.4,3.83,8,8.86])

# Error bars
throughput_error = np.array([2.5, 2.8, 4.5, 4.66, 4.7, 5.2, 6 ,7 , 4.99])
latency_error = np.array([0.3, 0.3, 0.3, 0.33, 0.35, 0.39, 0.41  , 0.58 , 0.62])


# Plot
fig, ax1 = plt.subplots(figsize=(9, 6))  # Set the figure size to 8x5 inches

# Plot throughput with error bars, using the index to create even spacing
index = np.arange(len(send_rate)) *3  # Index for uniform spacing
print(index)
ax1.bar(index, throughput, yerr=throughput_error, 
        color='#4567DB', label='Throughput objectiveRep', 
        capsize=0, width=0.6,  # Narrower width for cleaner look
        error_kw={'elinewidth': 1 , 'color':'#333333'}) 
ax1.set_xlabel('Send Rate (TPS)')
ax1.set_ylabel('Throughput (TPS)')
ax1.tick_params(axis='y')

# Set the x-ticks to the original send_rate values
ax1.set_xticks(index)
ax1.set_xticklabels(send_rate)  # Use original send_rate as labels

ax1.legend(loc='center left', bbox_to_anchor=(0.005, 0.94))

offset = 0.8  # Adjust the value to move the marker right
send_rate_offset = index + offset

# Create a second y-axis for latency
ax2 = ax1.twinx()

# Plot latency with markers and error bars, also using index for uniform spacing
ax2.plot(send_rate_offset, latency, color='#C2CFC2', marker='^', markersize=5, markeredgecolor="#333333", label='Latency objectiveRep')
ax2.errorbar(send_rate_offset, latency, yerr=latency_error, fmt='^', color='#333333', markersize=5, markeredgecolor="#333333", elinewidth=1)

# Set the y-axis limits for latency to range from 1 to 9
ax2.set_ylim(1, 10)
ax2.set_ylabel('Latency (s)')
ax2.tick_params(axis='y')

# Adjust legend for latency
ax2.legend(loc='center right', bbox_to_anchor=(0.805, 0.94))

# Title and layout
plt.title('Throughput and Latency vs Send Rate (Obj rep)')
fig.tight_layout()

# Save the figure to a file
plt.savefig("throughput_latency_objective_rep.png", dpi=300, bbox_inches='tight')  # Save as PNG with high resolution


# Show plot
plt.show()
```
