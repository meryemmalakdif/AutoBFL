```python
import matplotlib.pyplot as plt
import numpy as np

# Data
send_rate = np.array([50, 100, 150, 210, 260, 320, 360, 410])
throughput = np.array([45, 77, 118, 139, 166, 180, 175, 161])
latency = throughput * 20

# Error bars
throughput_error = np.array([3.6, 4.4, 5.5, 6.6, 7.4, 12.8, 10, 8.5])
latency_error = np.array([0.17, 0.17, 0.17, 0.17, 0.17, 0.25, 0.31, 0.35])

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot throughput with error bars
index = np.arange(len(send_rate)) * 3  # Index for uniform spacing
ax1.bar(index, throughput, 
        color='#FF8080', label='Average Throughput In Single-Layered BFL (TPS)', 
        capsize=0, width=0.6,  
        error_kw={'elinewidth': 1 , 'color':'#333333'}) 
legend = ax1.legend(loc='center left', bbox_to_anchor=(0.005, 0.98), fontsize=13)


# Set the x-ticks to the original send_rate values
ax1.set_xticks(index)
ax1.set_xticklabels(send_rate)  # Use original send_rate as labels
ax1.set_xlabel('Send Rate (TPS)', fontsize=14)
ax1.set_ylabel('Throughput (TPS)', fontsize=14)

# Plot latency as a second set of bars, slightly offset to the right
latency_index = index + 0.6  # Offset for latency bars
ax1.bar(latency_index, latency,
        color='#4567DB', label='Average Throughput In AutoBFL (TPS)', 
        capsize=0, width=0.6,  
        error_kw={'elinewidth': 1 , 'color':'#333333'}) 

# Customize ticks and legend
ax1.tick_params(axis='y')
legend = ax1.legend(loc='center left', bbox_to_anchor=(0.005, 0.935), fontsize=13)
legend.get_frame().set_linewidth(0)

# Title and layout
fig.tight_layout()
plt.savefig("./finalPlots/bc/comparaison.png", dpi=300, bbox_inches='tight')  # Save as PNG with high resolution


# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
send_rate = np.array([50, 90, 145, 200, 250, 320, 390, 420])
throughput = np.array([42.2, 81.3, 124.96, 165.56, 218.3, 284, 275, 259])
latency = np.array([1.536, 1.556, 1.5966, 1.606, 1.64, 3.17, 4.51, 6.6])

# Error bars
throughput_error = np.array([3.6, 4.4, 5.5, 6.6, 7.4, 12.8, 10, 8.5])
latency_error = np.array([0.17, 0.17, 0.17, 0.17, 0.17, 0.25, 0.31, 0.35])



# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))  # Set the figure size to 8x5 inches

# Plot throughput with error bars, using the index to create even spacing
index = np.arange(len(send_rate)) *3  # Index for uniform spacing
print(index)
ax1.bar(index, throughput, yerr=throughput_error, 
        color='#4567DB', label='Throughput submitLocalModel', 
        capsize=0, width=0.6,  # Narrower width for cleaner look
        error_kw={'elinewidth': 1 , 'color':'#333333'}) 
ax1.set_xlabel('Send Rate (TPS)', fontsize=14)
ax1.set_ylabel('Throughput (TPS)', fontsize=14)

ax1.tick_params(axis='y')

# Set the x-ticks to the original send_rate values
ax1.set_xticks(index)
ax1.set_xticklabels(send_rate)  # Use original send_rate as labels

legend = ax1.legend(loc='center left', bbox_to_anchor=(0.005, 0.97), fontsize=13)
legend.get_frame().set_linewidth(0)

offset = 0.8  # Adjust the value to move the marker right
send_rate_offset = index + offset

# Create a second y-axis for latency
ax2 = ax1.twinx()

# Plot latency with markers and error bars, also using index for uniform spacing
ax2.plot(send_rate_offset, latency, color='#C2CFC2', marker='^', markersize=5, markeredgecolor="#333333", label='Latency submitLocalModel')
ax2.errorbar(send_rate_offset, latency, yerr=latency_error, fmt='^', color='#333333', markersize=3, markeredgecolor="#333333", elinewidth=1)
plt.xticks(fontsize=11)  # Adjust the fontsize for x-axis tick labels
plt.yticks(fontsize=11)

# Set the y-axis limits for latency to range from 1 to 9
ax2.set_ylim(1, 8)
ax2.set_ylabel('Latency (s)', fontsize=14)
ax2.tick_params(axis='y')

# Adjust legend for latency
legend = ax2.legend(loc='center left', bbox_to_anchor=(0.005, 0.92),fontsize=13)

# Title and layout
# plt.title('Throughput and Latency vs Send Rate (submitModel)')
fig.tight_layout()

legend.get_frame().set_linewidth(0)
# Save the figure to a file
plt.savefig("./finalPlots/bc/throughput_latency_submitModel.png", dpi=300, bbox_inches='tight')  # Save as PNG with high resolution


# Show plot
plt.show()
```
